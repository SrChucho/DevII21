## Inequality, Control Rights, and Rent-Seeking

   Banerjee, Abhijit, Dilip Mookherjee, Kaivan Munshi, and Debraj Ray. 2001. “Inequality, Control Rights, and Rent-Seeking: Sugar Cooperatives in Maharashtra.” *Journal of Political Economy* 109 (1): 138–190.  ([link](https://debrajray.com/wp-content/uploads/2017/12/Sugar.pdf))

### A simpler model

The paper aims to model the distortionary effects of political capture in organizations.   

**Farmer's production decisions**

Farmers face a product price and choose a labor input $l$ to maximize profits given the sugar price $p$ and the market wage $w$. 
The farmer maximizes

$$
\pi(p) = \max_l p\cdot f(l) - w\cdot l
$$

If we assume $f(l) = 2 \cdot \sqrt l$ then the profit maximizing labor input can be easily shown to be:

$$
l^* = \frac{p^2}{w^2}
$$

If we also normalize the market wage to $w=1$, then $l^* = p^2$, optimal output is $q^* = 2 \cdot p$ and profits are $\pi(p)=p^2$

**Sugar Cooperatives**

Farmers sell their sugar to a sugar cooperative (which has the power to be a monopsony buyer in the area) at product price $p$.  The extent to which the coop will act as a rent-extracting monopsonist however depends on the balance of power within the cooperative -- whether decisions are balanced to reflect the interests of both big and small producers or not.

If the coop comes to be dominated by big farmers they may try to create and then extract monopsony rents. Now ideally, if you were to represent the big guys who have captured control of an organization you'd want to maintain efficient levels of production and then extract rents as efficiently as possible.  Often, however, mechanisms for efficient rent-extraction are blocked for incentive or political reasons.  Then those in power may resort to second-best mechanisms for rent extraction.

The 'power' of the elite group in the sugar cooperative is modeled parsimoniosly via a single parameter $\lambda$ that describes the relative weight that a 'social planner' (shorthand for the political mechanism determining policies in the coop) places on the welfare and interests of small farmers, compared to larger farmers.  

There are two policies that the coop decides that will affect resource allocation and rent distribution:  
1. the price $p$ paid to farmers for their raw sugar 
2. the lump sum payments $R_j$ paid to (or received from farmers of type $j \in \{S,B\}$ which stands for Small or Big. 

The coop may offer to buy sugar cane from farmers at a single common price $p$ and then resell the sugar at price $p^*>p$.  The full paper also models sugar cane processing costs but this doesn't really affect the main story so we'll ignore this here.  


Let's simplify a bit further here for the purposes of illustration by assuming that there is no difference between 'Big' and 'Small' farmers in terms of their numbers of extent of control over land.  The paper allows them to be different but then expresses things in per-capita terms which ends up making the expressions essentially similar to what we examine here.

The extent to which Big farmers exert control over decisions is captured by assuming that the principals of the sugar-cooperative act to maximize the following weighted social welfare function. 

$$
W = u^B + \lambda \cdot u^S
$$

where $u^B, u^S$ are, respectively, the welfare of the large and the small farmer 

$$
u^B = \pi^B(p) + R^B  \\
u^S = \pi^S(p) + R^S 
$$

and $\lambda$ is a welfare weight that determines how much weight is given to the welfare of the small.  If the big control the coop they will set $\lambda<1$.  

If the coop sets a price $p$ below the price $p^*$ at which the coop resells sugar, the coop will earn rents $(p^*-p)\cdot (q^B+q^S)$ which are distributed in lump-sum fashion to each group.

$$
R^B+R^S = (p^*-p)\cdot (q^B+q^S)
$$

Although not explicitly noted in the paper each farmer also has a participation constraint of the form:

$$u^B \ge0 \\
u^S \ge 0$$

Note that absent any further constraint this statement of the optimization problem allows $R^S$ to become negative - in effect it allows contracts where the small farmer must pay a fixed rent as a condition of participation.

**Proposition 1** (simplified):  The efficient way extract rent is to pay all classes $p=p^*$ to lead them to make the optimal labor allocation decisions and then extract the rent via efficient lump sum transfers.  

If $\lambda<1$ then the social planner will in effect push the small farmer down to their participation constraint:

$$
u^S = R^S + \pi(p^*)
$$

This implies the small farmer would be asked to surrender their profits which are then transferred to the big:
$$R^S = -\pi(p^*)  \\
R^B=\pi(p^*)
$$ 

**Efficient exploitation** of this sort hence keeps $p=p*$ to not distort the output decisions so as to make the pie as large as possible and then cuts the largest slice of the pie for the big farmer which is consistent with getting the small farmer to participate.

**Constraints on efficient rent-extraction**

Perhaps because rent extracton of this sort is so 'naked' and involves compelling the small farmer to hand over their profits, there are reasons why this this type of arrangement would not be observed much in practice. The get at this idea the authors impose the constraint:

$$R^S \ge 0$$

In this case it is easy to see that the big farmers will capture any rents  $R^B = (p^* - p) \cdot (q^S + q^B)$ but the 'social planner' faces a tradeoff.  As they lower the price $p<p^*$ this depresses the profits of both the small and the big farmers but it creates rents that can be captured by the big.  The more the big control the governance of the coop, the more they'll want to depress the price.  

The planner maximizes:

$$
W = \pi(p) + (p^* - p ) \cdot (2\cdot f(p))  + \lambda \cdot (\pi(p)) 
$$

Substituting in for optimized profits and output ($\pi(p)=p^2$ and $q^* = 2 \cdot p$) as derived above when $f(l) = 2 \cdot \sqrt l$ this can be written (after some regrouping) as:

$$
W = (1+ \lambda) \cdot  p^2 + (p^* - p )\cdot  4p
$$

Setting $\frac{dW}{dp}=0$ and solving for $p$ yields the optimal price:

$$
\hat p = \frac{2p^*}{3-\lambda}
$$

When $\lambda=1$ then it's optimal to set the price efficiently at $p=p^*$ and there is no output distortion.

When $\lambda<1$ then big farmers (who can capture the coop rents) have more influence and they will set $p<p^*$

For example as illustrated below, if $\lambda = 0.5$ the optimal $\hat p$ = 0.8.

At the efficient $p=p^*=1$ the farmers produce $q^B=q^S = 2$ each and the large farmer makes profit of 1.

When $\lambda = 1/2$ $\hat p=0.8$ and each farmer produces $q^B=q^S = 1.6$ each but the larger farmer is better off as they make :

$$
\pi(\hat p) + R^B =  0.64 + (1-0.8) \cdot 3.2 = 1.28
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
```

```python
def f(p):
    return 2*p

def profit(p):
    return p**2

def u(p,R):
    return profit(p) + R

def W(p, Rs, Rb, lmbda):
    return  u(p, Rb)  + lmbda * u(p, Rs)

def phat(pstar, lmbda):
    return 2*pstar/(3-lmbda)
```

```python
def plotW(lmbda):
    plt.figure(figsize=(8,6))
    pstar = 1
    popt = phat(pstar,lmbda)
    foo = f'{popt:5.3f}'
    pp = np.linspace(0.1, pstar, 20)
    rent = (pstar - pp)*(4*pp)
    W = profit(pp) + rent + lmbda * (profit(pp))
    plt.xlim(0,1)
    plt.ylim(0,2)
    plt.text(0.3,1.85,r'$\hat p = $'+  foo)
    plt.plot(pp, W, label='W')
    plt.plot(pp, profit(pp), label= r'$\pi(p)$')
    plt.plot(pp, profit(pp)+rent, label='rent')
    plt.plot(pp, f(pp), label=r'$q^S+q^B$' )
    plt.legend(loc='upper left', frameon=False)
    plt.axvline(popt, linestyle=':')
    plt.grid();
```

```python
plotW(1/2)
```

```python
interact(plotW, lmbda=(0.1,1,0.1));
```

### Notes on the original  model

Each cooperative has a fixed command area.  The production function is CRS in land and labor, but land area is fixed at the time of membership, so labor $l$ is the variable input.

$f(l)$  - Production function per acre (where $\l is labor per acre)

$S$ - land owned by small farmers

$B$ - land owned by large farmers

There are $M$ small and $N$ large farmers in an area of which $m$ small farmers and $n$ large farmers participate.

$\beta = \frac{m}{n}$  -- fraction of small to big growers participating

$\hat \beta = \frac{M}{N}$  -- fraction of small to big growers participating

$A=mS+nB$ acres out of $MS+NB=\hat A$ potential acres.

$c(K)$ -- variable cost is strictly decreasing in crushing capacity $K$

$G=G(K)$  -- a higher capacity entails higher setup cost

Each member is offered a price $p$ for sugarcane and maximizes:

$$
\pi(p)=p\cdot f(l) - w\cdot l
$$

#### Collective Decisions

The coop pays each member price $p$ per unit of sugar cane, it processes (crushes) the sugar cane and then sells processed sugar at market price $p^*$.


**Retained Earnings:**

$$
R=[p^*-c(K)-p]Q-G(K)
$$

These retained earnings are "diverted in a variety of ways that directly or indirectly benefit members."  To simplify assume the retained earnings $R$ are allocated between (per farmer) lump-sum transfers $R^B$ and $R^S$, where $R=mR^S+nR^B$

The collective chooses $p, R^B, R^S$.  The payoff to each type $T \in {B,S}$ is

$$
u^T=T\cdot \pi(p) + R^T
$$

Like a two-part tariff for each kind of farmer.  THe main institutional constraint are that the price $p$ be the same for both types of farmer and lump sum transfers to small farmers be non-negative:

$$
R^S \ge 0
$$

**Efficient level of capacity**

$q = p^* - c(K)$ -- price of sugar net of crushing costs.

$\sigma(q,p)=q\cdot f(l(p))-w\cdot l(p)$ -- social profit per acre

$A\cdot \sigma(q,p^*-c(K), p) - G(K)$  -- aggregate income of all farmers. Note that efficient value of $p=p^*-c(K)$

Efficient level of capacity should be chosen to maximize this aggregate income.

**Control Rights**

$$
W=u^B+\lambda u^S
$$

where $\lambda$ is the relaitve control rights of small growers vis a vis large.  

Assume $\lambda(\beta)$ is increasing in $\beta=\frac{m}{n}$ the relative number of small in the cooperative but make a *disproportionate control hypothesis*: 

$$\lambda(\beta) < \beta$$

*Proposition 1* follows immediately: $R^S=0$ "If small farmers enjoyed positive retained earnings, these earnings could be transferred from small to large growers at a per capita conversion rate of b, which is larger than the welfare weight of the small growers."

So the net payoff to being a small farmer is just the profit from cultivation, while the larger farmer gets a transfer as well:

$$u^S = S \pi(p)$$

$$u^B = B \pi(p) + R^B$$

Let $\rho(p,p^*; K) = [p^*-c(K)-p]f(l(p))$ be rent generated per unit acre of land devoted to sugar cane.  Then from the budget constraint for the cooperative:

$$
R^B=[p^*-c(K)-p](B+\beta \cdot S)f(l(p)) - \frac{G(K)}{n}
$$

This is made up of the social profit from the large grower's cane supply $\sigma(q,p^*-c(K), p)$ plus the grower's share of the rent generated on the small growers' supply $\beta [p^*-c(K)-p]Sf(l(p))$ less their share of setup costs $\frac{G(K)}{n}$

The large farmer's payoff is then:

$$
u^B = B\sigma(q,p^*-c(K), p)+\beta S \rho(p,p^*; K) - \frac{G(K)}{n}
$$

The second term is the rent extracted from cane delivered by small growers. It is maximized at the monopsony price.

In its absence the growers would set cane price p at its efficient level $p=p^*-c(K)$.

To capture rents they lower the cane price below the efficient level, trading off the loss of social profit on their own supply against the increased rents captured from cane supplied by the other group. 

The objective function of the (captured) cooperative is then:

$$
W= B\cdot  \sigma(q,p^*-c(K), p)+\beta \cdot S \cdot \rho(p,p^*; K) +\lambda (\beta)\cdot S \pi(p)   - \frac{G(K)}{n}
$$

The relative weights $\beta$ and $\lambda(\beta)$ respectively express the **rent-seeking effect** and the **control shift effect**

It's useful to express the above slightly differently:

$$
u^B = \left [ (B +\beta S) \cdot \sigma(q,p^*-c(K), p) - \frac{G(K)}{n} \right ] - \beta S \pi(p)
$$

And then 

$$
W= (B +\beta S) \cdot \sigma(q,p^*-c(K), p) -[\beta- \lambda (\beta)] \cdot S \pi(p)   - \frac{G(K)}{n}
$$

"The second term in this expression thus captures the entire source of divergence of the cooperative’s objective from social profit, resulting from the tension between the rent-seeking and control shift effects."
